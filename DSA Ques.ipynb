### Q1 Two Sum: Find two numbers that add to target.

```python
nums = [2, 7, 11, 15]
target = 9

def two_sum(nums: list, target: int) -> list:
    seen = {}

    for i in range(len(nums)):
        diff = target - nums[i]
        if diff in seen:
            return [seen[diff], i]
        seen[nums[i]] = i
    seen[nums[i]] = i


two_sum(nums, target)
```

### Q3 Pallindrome Check

```python
string = "madam "

def is_pallindrome(string:str) -> bool:
    lower_case  = string.lower()
    removing_space = lower_case.strip()
    reverse_string = removing_space[::-1]
    result = removing_space == reverse_string
    return result

print(is_pallindrome(string))
```

### Q4 Longest Substring Without Repeating Characters

### Longest Substring Without Repeating Characters

**Problem:**  
Given a string, find the **length** of the longest substring with **all unique characters**.

---

#### Key Points
- Substring = continuous sequence of characters  
- No character should repeat  
- Return **length**, not the substring

---

#### Example

Input: `"abcabcbb"`  
Output: `3`  
Explanation: `"abc"` is the longest substring without repeats

---

#### Intuition
- Traverse the string
- Keep a window of unique characters
- If duplicate appears → shrink window
- Track maximum length

---

#### Goal
Efficient solution using **Sliding Window + Set/HashMap** (O(n))

```python
def longest_substring(text: str) -> int:
    last_seen = {}
    start_index = 0
    longest_length = 0
    length_of_text = len(text)

    for current_index in range (length_of_text):

        current_char = text[current_index]

        if current_char in last_seen:
            start_index = max(start_index, last_seen[current_char] + 1)
    
        last_seen[current_char] = current_index
        longest_length = max(longest_length, current_index - start_index + 1)
    
    return longest_length

text = "abcabcbb"
result = longest_substring(text)
print(result)
```

### Q6 Move Zeroes

```python
sample_list = [3,4,0,17,6,0,9]

def move_zeroes(sample_list:list) -> list:
    result = []
    
    for number in  sample_list:
        if number !=0:
            result.append(number)

    length_of_result = len(result)
    length_of_sample_list = len(sample_list)
   
    while length_of_result < length_of_sample_list:
        result.append(0)
        length_of_result += 1

    return result

move_zero = move_zeroes(sample_list)
print(move_zero)
```

### Q7 Maximum Subarray (Kadane's Algorithm)

### Maximum Subarray (Kadane’s Algorithm)

**Problem:**  
Given an integer array `nums`, find the **maximum sum** of a **contiguous subarray**.

---

#### Key Points
- Subarray = continuous elements  
- Must contain at least one element  
- Return the **maximum possible sum**

---

#### Example

Input: `[-2,1,-3,4,-1,2,1,-5,4]`  
Output: `6`  
Explanation: Subarray `[4, -1, 2, 1]` has the largest sum

---

#### Goal
Efficient solution in **O(n)** time

```python
sample_list = [-2,1,-3,4,-1,2,1,-5,4]

def max_sub_array(sample_list: list) -> int:
    current_sum = 0
    max_sum = sample_list[0]

    for numb in sample_list:
        current_sum += numb

        if current_sum > max_sum:
            max_sum = current_sum

        elif current_sum < 0:
            current_sum = 0
    return max_sum


result = max_sub_array(sample_list)
print(result)
    
```

### Q8 Detect Cycle in Linked List

👉 Asked in AmEx interview

#### 🧠 Detect Cycle in Linked List (AmEx Interview)

#### 📌 Problem Statement
Given the head of a linked list, determine whether it contains a **cycle (loop)**.

A cycle exists if a node’s `next` pointer eventually points back to a previous node.

Example:


---

#### 🚀 Approach: Floyd’s Cycle Detection Algorithm (Tortoise & Hare)

Use two pointers:
- `slow` → moves 1 step at a time
- `fast` → moves 2 steps at a time

---

#### 💡 Key Idea
- If a cycle exists → fast and slow pointers will eventually meet
- If no cycle → fast reaches `None`

---


```python
class Node:
    def __init__(self, value):
        self.value = value
        self.next = None
    

def detect_cycle(head: Node) -> bool:
    first_pointer = head            # moves slowly (1 step)
    second_pointer = head           # moves fast (2 steps)

    while second_pointer and second_pointer.next:
        
        # move pointers
        first_pointer = first_pointer.next
        second_pointer = second_pointer.next.next

        # if they meet, cycle exists
        if first_pointer == second_pointer:
            return True
        
    # if loop ends, no cycle
    return False
```

```python
# Create nodes
a = Node(1)
b = Node(2)
c = Node(3)
d = Node(4)

# Connect nodes (1 → 2 → 3 → 4)
a.next = b
b.next = a
#c.next = d

# Test
print(detect_cycle(a))   # Expected: True
```
